```python
#Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('vendas_tecnologia.csv')


print("Primeiras 5 linhas do DataFrame:")
print(df.head())


print("\nInformações do DataFrame:")
df.info()
```

```python

print("Tipos de Dados e Contagem de Valores Não Nulos:")
df.info()

print("\nContagem de Valores Ausentes (NaN) por Coluna:")
print(df.isnull().sum())
```

```python

df['Data'] = pd.to_datetime(df['Data'])

print("\nTipos de Dados após conversão da 'Data_Venda':")
df.info()
```

```python
# Calcule as vendas totais
vendas_totais = df['Total_Venda'].sum()
print(f"Vendas Totais: R${vendas_totais:,.2f}")

# Calcule a média de venda por pedido
media_venda_por_pedido = df['Total_Venda'].mean()
print(f"Média de Venda por Pedido: R${media_venda_por_pedido:,.2f}")
```

```python
# Produtos mais vendidos por QUANTIDADE
produtos_mais_vendidos_qtd = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Produtos Mais Vendidos (Quantidade):\n", produtos_mais_vendidos_qtd)

# Produtos mais vendidos por VALOR (receita gerada)
produtos_mais_vendidos_valor = df.groupby('Produto')['Total_Venda'].sum().sort_values(ascending=False).head(5)
print("\nTop 5 Produtos Mais Vendidos (Valor):\n", produtos_mais_vendidos_valor)

# Visualização em gráfico de barras para produtos por valor
plt.figure(figsize=(10, 6))
# A linha ajustada para evitar o FutureWarning é esta abaixo:
sns.barplot(x=produtos_mais_vendidos_valor.index, y=produtos_mais_vendidos_valor.values,
            palette='viridis', hue=produtos_mais_vendidos_valor.index, legend=False)
plt.title('Top 5 Produtos Mais Vendidos por Receita')
plt.xlabel('Produto')
plt.ylabel('Receita Total (R$)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Categorias mais populares 
categorias_mais_populares = df['Categoria'].value_counts().head(5)
print("\nTop 5 Categorias Mais Populares:\n", categorias_mais_populares)

# Visualização em gráfico de barras
plt.figure(figsize=(10, 6))

sns.barplot(x=categorias_mais_populares.index, y=categorias_mais_populares.values,
            palette='magma', hue=categorias_mais_populares.index, legend=False)
plt.title('Top 5 Categorias Mais Populares (Número de Pedidos)')
plt.xlabel('Categoria')
plt.ylabel('Número de Pedidos')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Vendas por cidade
vendas_por_cidade = df.groupby('Cidade')['Total_Venda'].sum().sort_values(ascending=False)
print("\nVendas por Cidade:\n", vendas_por_cidade)

# Visualização em gráfico de barras
plt.figure(figsize=(10, 6))

sns.barplot(x=vendas_por_cidade.index, y=vendas_por_cidade.values, palette='cividis', hue=vendas_por_cidade.index, legend=False)
plt.title('Vendas Totais por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Total de Vendas (R$)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Vendas totais por Canal de Venda
vendas_por_canal = df.groupby('Canal_Venda')['Total_Venda'].sum().sort_values(ascending=False)
print("\nVendas Totais por Canal de Venda:\n", vendas_por_canal)

# Visualização em gráfico de barras
plt.figure(figsize=(8, 5))
sns.barplot(x=vendas_por_canal.index, y=vendas_por_canal.values,
            palette='Set2', hue=vendas_por_canal.index, legend=False)
plt.title('Vendas Totais por Canal de Venda')
plt.xlabel('Canal de Venda')
plt.ylabel('Total de Vendas (R$)')
plt.tight_layout()
plt.show()
```
